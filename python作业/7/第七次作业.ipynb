```python
import jieba
from wordcloud import WordCloud,ImageColorGenerator
from imageio import imread
import matplotlib.pyplot as plt
```

```python
#中文文档（水浒传）
```

```python
text=open("Shuihu.txt",'r',encoding="utf8").read()
for ch in "：“”''""‘’，。！；…\n\t《》":
    text=text.replace(ch," ")
word=jieba.lcut(text)
word=[w for w in word if len(w.strip())>1]
text=' '.join(word)
picture=imread('1.jpg')
wordcloud1=WordCloud(font_path='simhei.ttf',
 mask=picture,
 background_color='white',
 mode="RGB",
 max_font_size=100,
 min_font_size=20,
 max_words=200,
 random_state=None,
 prefer_horizontal=0.8,
 )
wordtemp=wordcloud1.generate(text)
im_color=ImageColorGenerator(picture)
wordcloud_img=wordtemp.recolor(color_func=im_color)

wcimg=wordtemp.to_image()
wcimg.save("3.jpg")
plt.imshow(wordcloud_img)
plt.show()
```

```python
#说明：从中文文档的词云图可看出：水浒传中宋江出现的次数最多，说明宋江是故事中的主要人物。
#除此之外也有“厮杀”，“性命”等词语频繁出现，说明水浒传是一部关于抗争的故事。
#“梁山泊”这一地名的频繁出现，可能说明这是文章的主要发生场所
#“卢俊义”，“武松”，“吴用”，“西门庆”，“晁盖”可能也是文章的主要配角。
#“且说”，“如何”等词，反映了该小说有一定的年代了
```

```python
text=open("English.txt",'r',encoding="utf8").read()
txt = text.lower()
for ch in '!"#$%&()*+,-./:;<=>?@[\\]^_‘{|}~':
 txt = txt.replace(ch, " ")
word = txt.split()
word=[w for w in word if len(w.strip())>1]
text=' '.join(word)
picture=imread('2.jpg')
wordcloud2=WordCloud(font_path='arial.ttf',
 mask=picture,
 background_color='white',
 mode="RGB",
 max_font_size=60,
 min_font_size=5,
 max_words=300,
 random_state=None,
 prefer_horizontal=0.8,
 )
wt=wordcloud2.generate(text)
im_color=ImageColorGenerator(picture)
wc_img=wt.recolor(color_func=im_color)
wcimg=wt.to_image()
wcimg.save("4.jpg")
plt.imshow(wc_img)
plt.show()
```

```python
#说明：通过本题的词云图，可看出文章中Chinese和America这两次出现的频率极高，说明文章是主要关于中美之间的对比行文。
#包括成都，和trump出现的次数也很多，说明成都可能是文章的重点叙事城市。
#包括amazon，customer的出现，可能说明这篇文章是关于网络购物和顾客的。
#包括pandemic的出现，说明可能是关于疫情方面的分析文章。
```
