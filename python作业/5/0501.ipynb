```python
import collections  #用collections中的counter进行词频统计
```

```python
f=open('The Rise of Made-in-China Diplomacy.txt','r',encoding='utf-8')
s=f.read()  #读取文件进入字符串
```

```python
d='\W+'  #正则表达式提取元素 用W+相当于^w+，筛选出所有单词
c=re.split(d,s)      
```

```python
e=collections.Counter(c)
print(e.most_common(20))   #统计出出现频率前20的单词
```
