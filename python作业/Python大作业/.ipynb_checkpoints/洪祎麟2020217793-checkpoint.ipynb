```python
def get_html(url,header):  
    try:          
        r=requests.get(url,headers=header) 
        r.raise_for_status() 
        r.encoding="utf-8"
        return r.text 
    except: 
        print("网页访问出错") 
        return None
```

```python
def get_Animedata(html):  
    lst=[] 
    soup=BeautifulSoup(html,"html.parser")
    s1=soup.find('div',class_="section") 
    s11=s1.find_all('div',class_="inner") 
    href1=[]
    for da in s11: 
        href=da.find('a')['href'] 
        if not href.startswith('http'): 
            href="".join(("https://bgm.tv",href)) 
        href1.append(href)
    pat1=r'<h3>.+?<a.+?>(.+?)</a>.+?</h3>'
    pat2=r'<h3>.+?<a.+?</a>(.+?)</h3>'
    pat3=r'<p class="info tip">\s+(.+?)</p>'
    pat4=r'.+?<small>Rank </small>(.+?)</span>'
    pat5=r'<p class="rateInfo">.+?"fade">(.+?)</small>.+?</p>'
    pat6=r'<p class="rateInfo">.+?"tip_j">\((.+?)\)</span>.+?</p>'
    pat7=r'<p class="info tip">\s+.+?/.+?/(.+?)\s+</p>'
    s11=str(s11)
    cname=re.findall(pat1,s11,re.S)
    jname=re.findall(pat2,s11,re.S)
    episode=re.findall(pat3,s11,re.S)
    time=re.findall(pat3,s11,re.S)
    director=re.findall(pat7,s11,re.S)
    rank=re.findall(pat4,s11,re.S)
    rate=re.findall(pat5,s11,re.S)
    number=re.findall(pat6,s11,re.S)
    for i in range(24):
        try:
            if jname[i]=='\n':
                if len(episode[i].split('/')[0])>=8:
                    lst.append([href1[i],cname[i],' ',' ',time[i],' ',rank[i],rate[i],number[i]])
                elif len(episode[i].split('/')[1])>=16:
                    lst.append([href1[i],cname[i],' ',episode[i].split('/')[0],time[i].split('/')[1],' ',rank[i],rate[i],number[i]])
                else:
                    lst.append([href1[i],cname[i],' ',episode[i].split('/')[0],time[i].split('/')[1],director[i],rank[i],rate[i],number[i]])
            else:
                if len(episode[i].split('/')[0])>=8:
                    lst.append([href1[i],cname[i],jname[i].split('<')[-2].split('>')[1],' ',time[i],' ',rank[i],rate[i],number[i]])
                elif len(episode[i].split('/')[1])>=16:
                    lst.append([href1[i],cname[i],jname[i].split('<')[-2].split('>')[1],episode[i].split('/')[0],time[i].split('/')[1],' ',rank[i],rate[i],number[i]])
                else:
                    lst.append([href1[i],cname[i],jname[i].split('<')[-2].split('>')[1],episode[i].split('/')[0],time[i].split('/')[1],director[i],rank[i],rate[i],number[i]])
        except:
            continue
    return lst
```

```python
def ClimbAnimedata():
    datalst=[]  
    for n in range(1,101): 
        url=f"https://bgm.tv/anime/browser?sort=rank&page={n}" 
        print(f"正在爬取bangumi的动画排行榜，当前在第{n}页")  
        html=get_html(url,header) 
        if html is None: 
            continue
        lst=get_Animedata(html) 
        datalst.extend(lst) 
    print("爬取完成啦！")   
    return datalst 
```

```python
def stopwordslist():
    stopwords = [line.strip() for line in open('stopwords.txt',encoding='UTF-8').readlines()]
    return stopwords
```

```python
def Fenci(sentence):
    sentence_depart = jieba.cut(sentence.strip())
    stopwords = stopwordslist()
    outstr = ''
    for word in sentence_depart:
        if word not in stopwords:
            if word != '\t':
                outstr += word
                outstr += " "
    return outstr
```

```python
def get_AnimeCommentData(html):  
    lst=[] 
    soup=BeautifulSoup(html,"html.parser")
    s1=soup.find_all('div',class_="item clearit") 
    
    
    pat1=r'.+?<div class="text">.+?">(.+?)</a>'
    pat2=r'.+?@(.+?)</small>'
    pat3=r'.+?<span class="starlight stars(.+?)"></span>'
    pat4=r'.+?<p>(.+?)</p>'
    pat5=r'.+?@.+?</small>(.+?)<p>'

    s1=str(s1)
    
    nickname=re.findall(pat1,s1,re.S)
    time=re.findall(pat2,s1,re.S)
    rate1=re.findall(pat3,s1,re.S)
    rate2=re.findall(pat5,s1,re.S)
    comment=re.findall(pat4,s1,re.S)


    
    for i in range(20):
        try:
            if len(rate2[i])==1:
                rate1.insert(i,' ')
            lst.append([nickname[i],time[i],rate1[i],comment[i]])
        except:
            continue
    return lst
```

```python
def AnimeCommentAnalysis2(AnimeCommentList2):
    AnimeComment=str(list(AnimeCommentList2['评论']))
    text=Fenci(AnimeComment)

    picture=imread('Source2.png')
    wordcloud1=WordCloud(font_path='simhei.ttf',
     mask=picture,
     background_color='white',
     mode="RGB",
     max_words=1000,
     random_state=None,
     prefer_horizontal=0.8,
     )
    wordtemp=wordcloud1.generate(text)
    im_color=ImageColorGenerator(picture)
    wordcloud_img=wordtemp.recolor(color_func=im_color)

    wcimg=wordtemp.to_image()
    wcimg.save("AnimeWordCloud2.png")

    AnimeCommentList2['时间']=AnimeCommentList2['时间'].str.replace('-','年')
    
    
    
    Num_Year=[0]*4
    Rate_Year=[0]*4
    AvgRate=[0]*4
    
    for i in range(10,1420):
        try:
            if AnimeCommentList2['分数'][i]==' ':
                AnimeCommentList2['分数'][i]='8.5'         
            elif int(AnimeCommentList2['时间'].str.split('年')[i][0])==2018:
                Num_Year[0]=Num_Year[0]+1
                Rate_Year[0]=Rate_Year[0]+eval(AnimeCommentList2['分数'][i])         
            elif int(AnimeCommentList2['时间'].str.split('年')[i][0])==2019:
                Num_Year[1]=Num_Year[1]+1
                Rate_Year[1]=Rate_Year[1]+eval(AnimeCommentList2['分数'][i])          
            elif int(AnimeCommentList2['时间'].str.split('年')[i][0])==2020:
                Num_Year[2]=Num_Year[2]+1
                Rate_Year[2]=Rate_Year[2]+eval(AnimeCommentList2['分数'][i])         
            elif int(AnimeCommentList2['时间'].str.split('年')[i][0])==2021:
                Num_Year[3]=Num_Year[3]+1
                Rate_Year[3]=Rate_Year[3]+eval(AnimeCommentList2['分数'][i])          
        except:
            continue
                  
    Num_Year[0]=Num_Year[0]+10  
    Rate_Year[0]=Rate_Year[0]+85
    
    
    for i in range(4):
        AvgRate[i]=round(Rate_Year[i]/Num_Year[i],2)
    Period = ['2018年','2019年','2020年','2021年']
    v7 = Num_Year
    v8 = AvgRate

    bar = Bar('2018-2021年间摇曳露营动画每年所发表评论的人数', title_pos='center', width=900)
    bar.add('',Period,v7,is_stack=True,mark_point=["max",'min'],mark_line=["average"])
    line = Line()
    line.add('',Period,v8,yaxis_min = 8,mark_point=["max",'min'],mark_line=["average"])

    overlap = Overlap()
    overlap.add(bar,is_add_yaxis = True)
    overlap.add(line,yaxis_index = 1,is_add_yaxis = True)
    overlap.render("CommentCompare2.html")
```

```python
def DataClean(datadict1):
    lst3=list(datadict1.keys())
    len3=len(lst3)
    for i in range(0,len3):
        try:
            for j in range(i+1,len3):   
                if lst3[i][:5]==lst3[j][:5]:  
                    if(len(lst3[i])<len(lst3[j])):
                        datadict1[lst3[i]]=datadict1[lst3[j]]+datadict1[lst3[i]]
                        del datadict1[lst3[j]]
                        del lst3[j]
                        len3=len(lst3)
                    else:
                        datadict1[lst3[j]]=datadict1[lst3[j]]+datadict1[lst3[i]]
                        del datadict1[lst3[i]]
                        del lst3[i]
                        len3=len(lst3)
                if j ==len3-1:
                    break
            if i ==len3-1:
                break
        except:
            continue
    try:
        datadict1['吹响！悠风号']=datadict1['吹响！悠风号']+datadict1['剧场版 吹响！悠风号～誓言的终章～']
        datadict1['魔法少女小圆']=datadict1['魔法少女小圆']+datadict1['剧场版 魔法少女小圆 [新篇] 叛逆的物语']
        datadict1['蜂蜜与四叶草']=datadict1['蜂蜜与四叶草II']+datadict1['蜂蜜与四叶草']
        datadict1['AIR']=datadict1['AIR']+datadict1['AIR 夏季篇']
        del datadict1['剧场版 吹响！悠风号～誓言的终章～']
        del datadict1['剧场版 魔法少女小圆 [新篇] 叛逆的物语']
        del datadict1['蜂蜜与四叶草II']
        del datadict1['AIR 夏季篇']
        print("数据清理完成啦！")
    except:        
        print("数据清理完成啦！")
    return datadict1
```

```python
def AnimeRecommendation1(lst1,lst2):
    max1=lst1[0][1]
    max2=lst2[0][1]
    min1=1
    min2=1
    datadict1=dict(lst1)
    datadict2=dict(lst2)
    for v,k in datadict1.items():
        datadict1[v]=(k-1)/(max1-1)
    for v,k in datadict2.items():
        datadict2[v]=(k-1)/(max2-1) 
    for v,k in datadict2.items():
        if v in datadict1.keys():
            datadict1[v]=round((datadict1[v]+k)/2,2)
        else:
            datadict1[v]=round(k/2,2)
    datadict1=sorted(datadict1.items(), key=lambda x:x[1], reverse=True)
    lst1=[y[0] for y in datadict1[:11]]
    lst2=[y[1] for y in datadict1[:11]]
    bar = Bar('看过《CLANNAD ～AFTER STORY～》还推荐看的动画',title_pos='center',width=1500,height=650)
    bar.add('',lst1,lst2,is_stack=True,xaxis_label_textsize=12,xaxis_rotate=45)
    grid = Grid(width=1500,height=650)
    grid.add(bar, grid_bottom="25%")
    grid.render('AnimeRecommendation1.html')


    df = pd.DataFrame(datadict1)
    df.to_excel('AnimeRecommendation1.xlsx')
    return datadict1
```
