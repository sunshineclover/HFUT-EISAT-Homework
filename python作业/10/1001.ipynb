```python
import requests  
from bs4 import BeautifulSoup
import pandas as pd   #引用本题所需要的库
```

```python
def get_html(url):  #定义函数 该函数用于获取网页里面的html信息
    try:          #try except异常处理机制，若失败则直接返回“网页访问出错”，防止代码直接崩溃
        r=requests.get(url)  #用requests的get方法获取网址
        r.raise_for_status() #用于判断访问网络的状态（正常即为200）
        r.encoding=r.apparent_encoding #改变提取出的编码为正确的编码形式
        return r.text #将从网页提取出的东西返回为文本文件
    except: #try except异常处理机制同上。此处为异常出现的情况
        print("网页访问出错") #则输出访问出错，返回None
        return None
```

```python
def get_pdata(html):  #定义函数 该函数通过使用BeautifulSoup4库，筛选出所需要的值并返回值列表中
    lst=[] #先定义一个空列表，之后用于存储数据
    soup=BeautifulSoup(html,"html.parser") #先将BeautifulSoup的解析器设置为html.parser
    s1=soup.find('ul',class_="news_list") #再在其中挑选出'ul且class="new_list"的部分'，即为筛选出新闻标题，时间，超链接所在的部分
    s11=s1.find_all('li') #筛选出所在的'li'中，之后在其中进行提取。
    for da in s11: #for循环开始遍历所有的<li>
        href=da.find('a')['href'] #提取出超链接
        if not href.startswith('http'): #如果提取的超链接不全（即为缺少前面一部分），则需要通过人为手工添加
            href="".join(("http://kyy.hfut.edu.cn",href)) #这里使用字符串的.join方法进行添加
        date=da.find('b').string #提取出时间
        name=da.find('p').string #提取出标题
        lst.append([name,href,date]) #将以上三个提取出的元素一同放入之前建的lst这一列表中
    return lst #返回该列表便于之后使用
```

```python
def climbdata(): #定义函数，该函数用于爬取网页的所有页数来爬取数据
    datalst=[]  #定义一个空列表，之后用于存储数据
    for n in range(1,136): #遍历循环，遍历所有网页的所有页，便于爬取全部数据
        url=f"http://kyy.hfut.edu.cn/tzgg/list{n}.htm" #需要爬取数据的网页，用{n}即是为了通过循环来爬取每一页
        print(f"现在已经下载到：第{n}页了")  #输出正在下载页数的标识
        html=get_html(url) #通过之前的函数，获得具体某一网页的全部文本
        if html is None: #如果网页没有东西，直接访问下一页
            continue
        lst=get_pdata(html) #通过运用这一函数，筛选出所需的三项数据。
        datalst.extend(lst) #把之前筛选出的数据存入datalst之后（注意需要使用extend方法接在后面）
    print("下载结束啦！")   #全部下载完毕后输出标识
    return datalst #返回存储数据的列表
```

```python
lst=climbdata()  #用lst接受之前函数最后返回的列表（此时列表中已经全为所需的数据）
df=pd.DataFrame(lst,columns=['新闻标题','超链接','发布时间']) #将列表转换为DataFrame格式
df.to_excel('1001.xlsx')  #将DataFrame格式最后转化为xlsx格式输出
```
